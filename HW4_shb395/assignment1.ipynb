```python
import numpy as np
import matplotlib as plt
%pylab inline
```

```python
# Get 100 random numbers between 10 and 2000
np.random.seed(395)
samples = np.random.randint(10,2000,100)
```

```python
print(samples)
```

```python
mean = 5
samples_len = len(samples)
```

```python
# Normal distribution
np.random.seed(395)
nums_list = []

for i in range(samples_len):
    numbers = np.random.normal(loc=mean, size=samples[i])
    nums_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(samples_len):
    ax.plot(nums_list[i].size, nums_list[i].mean(), '.')
```

```python
# Poisson
np.random.seed(395)
nums_list = []

for i in range(samples_len):
    numbers = np.random.poisson(lam=mean, size=samples[i])
    nums_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(samples_len):
    ax.plot(nums_list[i].size, nums_list[i].mean(), '.')
```

```python
# Binomial
np.random.seed(395)
nums_list = []

# Set the probability at .5 just so that the data resembles a common binomial distribution, such as flipping a coin
for i in range(samples_len):
    numbers = np.random.binomial(mean *2, p=.5, size=samples[i])
    nums_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(samples_len):
    ax.plot(nums_list[i].size, nums_list[i].mean(), '.')
```

```python
# Chi squared
np.random.seed(395)
nums_list = []

for i in range(samples_len):
    numbers = np.random.chisquare(5, size = samples[i])
    nums_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(samples_len):
    ax.plot(nums_list[i].size, nums_list[i].mean(), '.')
```

```python
# Logistic distribution
np.random.seed(395)
nums_list = []

for i in range(samples_len):
    numbers = np.random.logistic(loc=mean, size=samples[i])
    nums_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(samples_len):
    ax.plot(nums_list[i].size, nums_list[i].mean(), '.')
```
