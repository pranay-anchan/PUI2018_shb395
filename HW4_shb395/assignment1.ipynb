```python
# Import libraries
import numpy as np
import matplotlib as plt
import pylab as pl
%pylab inline
```

```python
# Define a sample generator function to create a new set of 100 random numbers between 10 and 2000 
# for each distribution
np.random.seed(395)

def sample_generator():
    samples = np.random.randint(10,2000,100)
    return samples
```

```python
# Set the mean and create an open list to collect all the sample means for each distribution
mean = 1
all_means = []
```

```python
# Normal
np.random.seed(395)

norm_samples = sample_generator()    

norm_list = []
norm_means = []

for i in range(len(norm_samples)):
    numbers = np.random.normal(loc=mean, size=norm_samples[i])
    norm_list.append(numbers)
    
ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(len(norm_samples)):
    all_means.append(norm_list[i].mean())
    norm_means.append(norm_list[i].mean())
    ax.plot(norm_list[i].size, norm_list[i].mean(), '.')

ax.set_title('Normal Distribution', fontsize = 15)
ax.set_ylabel('Mean', fontsize = 12)
ax.set_xlabel('Number of samples', fontsize = 12)
```

### Normal - CLT

Unsurprisingly, the plot of normally distributed data exhibits a relatively clear bell shaped curve, which supports the Central Limit Theorem.

```python
# Poisson
np.random.seed(395)

pois_samples = sample_generator()

pois_list = []
pois_means = []

for i in range(len(pois_samples)):
    numbers = np.random.poisson(lam=mean, size=pois_samples[i])
    pois_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(len(pois_samples)):
    all_means.append(pois_list[i].mean())
    pois_means.append(pois_list[i].mean())
    ax.plot(pois_list[i].size, pois_list[i].mean(), '.')

ax.set_title('Poisson Distribution', fontsize = 15)
ax.set_ylabel('Mean', fontsize = 12)
ax.set_xlabel('Number of samples', fontsize = 12)
```

### Poisson - CLT

The plot of the Poisson distribution resembles the bell shaped curve of a normal distribution, as predicted by the Central Limit Theorem. But because the standard deviation of a Poisson distribution is equal to the mean (which, in this case, is 1), the distribution is much narrower than the normal distribution seen above.

```python
# Binomial
np.random.seed(395)

bi_samples = sample_generator()

bi_list = []
bi_means = []

# Set the probability at .5 just so that the data resembles a common binomial distribution, such as flipping a coin
for i in range(len(bi_samples)):
    numbers = np.random.binomial(mean *2, p=.5, size=bi_samples[i])
    bi_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(len(bi_samples)):
    all_means.append(bi_list[i].mean())
    bi_means.append(bi_list[i].mean())
    ax.plot(bi_list[i].size, bi_list[i].mean(), '.')
    
ax.set_title('Binomial Distribution', fontsize = 15)
ax.set_ylabel('Mean', fontsize = 12)
ax.set_xlabel('Number of samples', fontsize = 12)
```

### Binomial - CLT

The binomial distribution resembles the bell shaped curve of the normal distribution, as predicted by the Central Limit Theorem, but with a broader peak at the mean.

```python
# Chi squared
np.random.seed(395)

chi_samples = sample_generator()

chi_list = []
chi_means = []

for i in range(len(chi_samples)):
    numbers = np.random.chisquare(1, size = chi_samples[i])
    chi_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(len(chi_samples)):
    all_means.append(chi_list[i].mean())
    chi_means.append(chi_list[i].mean())
    ax.plot(chi_list[i].size, chi_list[i].mean(), '.')
    
ax.set_title('Chi Squared Distribution', fontsize = 15)
ax.set_ylabel('Mean', fontsize = 12)
ax.set_xlabel('Number of samples', fontsize = 12)
```

### Chi Squared - CLT

The chi squared distribution resembles the bell shaped curve of the normal distribution, as predicted by the Central Limit Theorem.

```python
# Logistic distribution
np.random.seed(395)

log_samples = sample_generator()

log_list = []
log_means = []

for i in range(len(log_samples)):
    numbers = np.random.logistic(loc=mean, size=log_samples[i])
    log_list.append(numbers)

ax = plt.figure(figsize=(5,5)).add_subplot(111)
for i in range(len(log_samples)):
    all_means.append(log_list[i].mean())
    log_means.append(log_list[i].mean())
    ax.plot(log_list[i].size, log_list[i].mean(), '.')

ax.set_title('Logistic Distribution', fontsize = 15)
ax.set_ylabel('Mean', fontsize = 12)
ax.set_xlabel('Number of samples', fontsize = 12)
```

### Logistic - CLT

The logistic distribution resembles the bell shaped curve of the normal distribution, as predicted by the Central Limit Theorem.

```python
# Plot the sample means from one list as one color
pl.figure(figsize=(10, 10))
pl.hist(all_means,bins=30)
pl.xlabel('Sample mean', fontsize = 18)
pl.ylabel('Number', fontsize = 18)
pl.title('All Sample Means', fontsize = 25)
```

### Sample means - CLT

The histogram of sample means, collectively, clearly resembles a normal distribution as predicted by the Central Limit Theorem.

```python
# Plot the sample means as separate colors
pl.figure(figsize=(10, 10))
pl.hist(norm_means, alpha = .5, bins=30)
pl.hist(pois_means, alpha = .5, bins=30)
pl.hist(bi_means, alpha = .5, bins=30)
pl.hist(chi_means, alpha = .5, bins=30)
pl.hist(log_means, alpha = .5, bins=30)
pl.xlabel('Sample mean', fontsize = 18)
pl.ylabel('Number', fontsize = 18)
pl.title('All Sample Means', fontsize = 25)
```
