```python
import os
import sys

# you can use the numpy library for random number generation: 
# http://docs.scipy.org/doc/numpy/reference/routines.random.html

import numpy as np
```

```python
# code here...
np.random.seed(1)
```

```python
ReprRandAll = np.zeros((3, 2, 100))
print("ReprRandAll:")
print(ReprRandAll)
print("")

mean1, mean2, sigma = 10, 20, 5
for i in range(3):
    newReprRand1 = mean1 + sigma*np.random.randn(100)
    newReprRand2 = mean2 + sigma*np.random.randn(100)
    ReprRandAll[i,0,:] = newReprRand1
    ReprRandAll[i,1,:] = newReprRand2
    print("new xs:")
    print(newReprRand1)
    print("")
    print("ReprRandAll with new xs:")
    print(ReprRandAll[i,0,:])
    print("")
    print("new ys:")
    print(newReprRand2)
    print("")
    print("ReprRandAll with new ys:")
    print(ReprRandAll[i,1,:])
    print("")
    i += 1

print("New whole ReprRandAll:")
print(ReprRandAll)
```
