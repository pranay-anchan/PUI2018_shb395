### Tennis facility data acquisition

```python
url = "https://www.nycgovparks.org/bigapps/DPR_Tennis_001.json"
resp = requests.request('GET', url)
tennis = json.loads(resp.text)
```

```python
name, lat, lng, crt_num = [], [], [], []
for i in range(len(tennis)):
    name.append(tennis[i]['Name'])
    crt_num.append(tennis[i]['Courts'])
    lat.append(tennis[i]['lat'])
    lng.append(tennis[i]['lon'])

tennisdf = pd.DataFrame([name,crt_num,lat,lng]).T
```

```python
tennisdf.head(5)
```

```python
tennisdf.shape
```

```python
# Number of facilities
print("Number of public tennis facilities in NYC: " + str(len(tennisdf)))
```

```python
tennisdf.rename(columns={0:'name', 1:'crt_num', 2:'lat', 3:'lon'}, inplace=True)
# Drop rows that have nans in either the lat or long columns
tennisdf.dropna(subset=['lat','lon'], inplace=True)
```

```python
# Number of facilities after dropping nans
print("Number of public tennis facilities in NYC after removing facilities missing lat \
long info: " + str(len(tennisdf)))
```

```python
tennisdf.head(5)
```

```python
tennisdf.reset_index(drop=True, inplace=True)
```

```python
tennisdf['lat'] = tennisdf['lat'].astype(float)
tennisdf['lon'] = tennisdf['lon'].astype(float)
```

```python
tennisdf['lonlat'] = list(zip(tennisdf.lon, tennisdf.lat))
```

```python
tennisdf['geometry'] = tennisdf[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
tennisdf.head(5)
```

```python
tennisdf['crt_num'] = tennisdf['crt_num'].astype(int)
```

```python
tennisgdf = gpd.GeoDataFrame(tennisdf)
tennisgdf.crs = from_epsg(4326)
```

```python
tennisgdf.head(5)
```

```python
# Plot of tennis facility locations
# Notice that there's an outlier
f, ax = plt.subplots(figsize=(15,15))
tennisgdf.plot(alpha=1, ax=ax)
plt.title("NYC Zip Codes", fontsize=20)
plt.show()
```

```python
# Locate outlier
print(tennisgdf['lon'].max())
```

```python
# Remove outlier
tennisgdf = tennisgdf[tennisgdf.lon < 0]
```

```python
tennisgdf = tennisgdf.to_crs(epsg=2263)
zipshape = zipshape.to_crs(epsg=2263)
```

### Plotting

```python
# Plot of NYC zip codes with tennis facility locations
f, ax = plt.subplots(figsize=(15,15))
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
zipshape.plot(alpha=1, linewidth=0.5, ax=ax, color='w', edgecolor='black')
tennisgdf.plot(alpha=1, ax=ax)
plt.title("Distribution of Public Tennis Facilities in NYC", fontsize=20)
plt.show()
```

## Note on above plot

The plot above shows the placement of all public tennis courts (with latitude and longitude information available on NYC Open Data) mapped on top of NYC zip codes. Note that not all public tennis facilities are mapped since the latitude and longitude information for some facilities was not available on NYC Open Data (21 out of the total 81 facilities were dropped since they were missing latitude and longitude information).

```python
merged = gpd.sjoin(tennisgdf,zipshape)
```

```python
crts_pp = merged.groupby('ZIPCODE')['crt_num'].agg(np.sum).reset_index(name='crt_num')
```

```python
crts_pp = crts_pp.merge(zipshape[['ZIPCODE','POPULATION']], how='left', on='ZIPCODE')
```

```python
crts_pp['ZIPCODE'] = crts_pp['ZIPCODE'].astype(int)
crts_pp['crts_pp'] = crts_pp['crt_num']/crts_pp['POPULATION']
zipshape['ZIPCODE'] = zipshape['ZIPCODE'].astype(int)
zipshape_merge = zipshape.merge(crts_pp[['ZIPCODE','crts_pp']], how='left', on='ZIPCODE')
```

```python
zipshape_merge['crts_pp'].fillna(0, inplace=True)
```

```python
zipshape_merge['crts_pp'] = zipshape_merge['crts_pp']*1000
```

```python
# Plot of NYC zip codes with tennis facility locations
f, ax = plt.subplots(figsize=(15,15))
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
zipshape_merge.plot(column="crts_pp", alpha=.8, linewidth=0.5, ax=ax, label="crts_pp", \
              scheme="Equal_interval", k=10, legend=True)
tennisgdf.plot(alpha=1, ax=ax, markersize=75)
leg = ax.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
plt.title("Heatmap Distribution of Public Tennis Facilities in NYC", fontsize=20)
plt.show()
```

## Note on above plot

The plot above is a heatmap showing which NYC zip codes provide residents with more or less access to public tennis courts. Darker colored zip codes have the lowest access while lighter colors indicate zip codes with relatively more access. Note that the legend has been scaled up by a factor of $10^{3}$. So each gradient should be multiplied by 1/1000 to acquire the actual figures for the number of tennis courts per person in a given zip code. (I would have used scientific notation for the legend but could not find a method to make this modification.)

As in the previous plot, not all public tennis facilities are mapped since the latitude and longitude information for some facilities was not available on NYC Open Data (21 out of the total 81 facilities were dropped since they were missing latitude and longitude information).

The plot shows that overall, many zip codes in NYC have no access to public tennis facilities.
